```python
import pandas as pd
```

```python
# Read in file
df = pd.read_csv("Resources/2019-citibike-all2.csv")
df.head()
```

```python
# Select desired columns by index
df2 = df.iloc[:,3:12]
df2.head()
```

```python
# Add Ride ID column to data frame
df2["rideid"] = df["rideid"]
df2.head()
```

```python
# Make new column names
columns = ["Start/End", "Ride ID", "Station ID", "Station Name", "Latitude", "Longitude", "Bike ID"]
```

```python
# Create empty data frames to store info
df3 = pd.DataFrame(columns=columns)
df4 = pd.DataFrame(columns=columns)
```

```python
# Populate df3 with 'ride start' info
df3["Ride ID"] = df2["rideid"]
df3["Station ID"] = df2["start station id"]
df3["Station Name"] = df2["start station name"]
df3["Latitude"] = df2["start station latitude"]
df3["Longitude"] = df2["start station longitude"]
df3["Bike ID"] = df2["bikeid"]
```

```python
df3["Start/End"] = "Start"
df3.tail()
```

```python
# Populate df4 with 'ride end' info
df4["Ride ID"] = df2["rideid"]
df4["Station ID"] = df2["start station id"]
df4["Station Name"] = df2["start station name"]
df4["Latitude"] = df2["start station latitude"]
df4["Longitude"] = df2["start station longitude"]
df4["Bike ID"] = df2["bikeid"]
```

```python
df4["Start/End"] = "End"
df4.tail()
```

```python
df5 = pd.concat([df3, df4])
df5.tail()
```

```python
df5 = df5.sort_values(by=["Ride ID"])
df5.head()
```

```python
df6 = df5.reset_index(drop=True)
df6
```

```python
df6.to_csv("Resources/ride-paths.csv", index=False)
```
